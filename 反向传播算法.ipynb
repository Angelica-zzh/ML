```python

"""
实现反向传播
"""
import torch

x_data = [1.0,2.0,3.0]
y_data = [2.0,4.0,6.0]
# 定义tensor,创建权重
W = torch.Tensor([1.0])
# 设置计算梯度，默认为False
W.requires_grad = True
# 前馈函数
def forward(x,w):
    # 这个时tensor的数乘
    return x*w
# 损失函数
def loss(x,y,w):
    y_pred = forward(x,w)
    return (y_pred-y)**2

# 迭代100次，更新权值
for epoch in range(100):
    for x,y in zip(x_data,y_data):
        # 计算损失，返回的是一个tensor
        los = loss(x,y,W)
        # 反向传播
        los.backward()
        print('\tgrad:',x,y,W.grad.item())
        # 更新权重
        # 这里的grad也是一个tensor，需要获取数值
        W.data = W-0.01*W.grad.data
        # 梯度清零
        W.grad.data.zero_()
    print("Epoch:",epoch+1,los.item())
    print(1)

```
